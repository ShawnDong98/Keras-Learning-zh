## 3.1 What’s TensorFlow?

TensorFlow是一个基于Python的免费开源机器学习平台，主要由谷歌开发。与NumPy非常相似，TensorFlow的主要目的是使工程师和研究人员能够在数值张量上操作数学表达式。但TensorFlow在以下方面远远超出了NumPy的范围：

- 它可以自动计算任何可微表达式的梯度（如第2章所示），使其非常适合机器学习。
- 它不仅可以在CPU上运行，还可以在GPU和TPU上运行，这些高度并行的硬件加速器。
- TensorFlow中定义的计算可以很容易地分布在许多机器上。
- TensorFlow程序可以导出到其他运行时，如C++、JavaScript（适用于基于浏览器的应用程序）或TFLite（适用于在移动设备或嵌入式设备上运行的应用程序）等。这使得TensorFlow应用程序在实际环境中易于部署。

重要的是要记住，TensorFlow不仅仅是一个库。它实际上是一个平台，拥有一个庞大的组件生态系统，其中一些由谷歌开发，一些由第三方开发。例如，有用于强化学习研究的TF-Agents，有用于行业强度机器学习工作流程管理的TFX，有用于生产部署的TF-Serving，有预训练模型的TF-Hub存储库......这些组件一起涵盖了非常广泛的用例，从尖端研究到大规模生产应用。

TensorFlow的规模相当不错：例如，Oak Ridge National Lab 的科学家用它来在IBM Summit超级计算机的27,000个GPU上训练1.1 ExaFLOP极端天气预报模型。同样，谷歌使用TensorFlow开发了非常计算密集型的深度学习应用程序，如国际象棋和围棋代理AlphaZero。对于您自己的型号，如果您有预算，您可以现实地希望在小型TPU Pod或谷歌云或AWS上租用的大型GPU集群上扩展到大约10个PetaFLOP。这仍将是2019年顶级超级计算机峰值计算能力的1%左右！

## What’s Keras?

Keras是Python的深度学习API，构建在TensorFlow之上，为定义和训练任何类型的深度学习模型提供了便捷的方式。Keras最初是为了研究而开发的，目的是实现快速的深度学习实验。

通过TensorFlow，Keras可以在顶级不同类型的硬件（见图1）上运行——GPU、TPU或普通CPU——并可以无缝扩展到数千台机器。



Keras以优先考虑开发人员体验而闻名。它是为人类而不是为机器设计的 API。 它遵循减少认知负担的最佳做法：它提供一致和简单的工作流程，最大限度地减少常见用例所需的操作数量，并在用户错误时提供清晰和可操作的反馈。这使得Keras作为初学者易于学习，作为专家使用效率很高。

截至2019年底，Keras拥有超过37万名用户，从初创企业和大公司的学术研究人员、工程师和数据科学家，到研究生和爱好者。Keras用于谷歌、Netflix、优步、欧洲核子研究组织、美国宇航局、Yelp、Instacart、Square和数百家初创公司，致力于解决每个行业的广泛问题。Keras也是机器学习竞赛网站Kaggle上一个受欢迎的框架，在那里，大多数深度学习比赛都是使用Keras模型赢得的。

## 3.3 Keras and TensorFlow: a brief history

Keras比TensorFlow早8个月。它于2015年3月发布，TensorFlow于2015年11月发布。您可能会问，如果Keras建立在TensorFlow之上，那么在TensorFlow发布之前，它如何存在？Keras最初建在Theano之上，Theano是另一个提供自动微分和GPU支持的张量操作库——这是同类中最早的。Theano是在蒙特利尔大学蒙特利尔学习算法研究所（MILA）开发的，在许多方面是TensorFlow的前身。它率先提出了使用静态计算图进行自动微分和将代码编译到CPU和GPU的想法。

2015年底，在TensorFlow发布后，Keras被重构为多后端架构：可以将Keras与Theano或TensorFlow一起使用，两者之间的切换就像更改环境变量一样简单。到2016年9月，TensorFlow已达到技术成熟度，使其成为Keras的默认后端选项。2017年，Keras增加了两个新的后端选项：CNTK（由微软开发）和MXNet（由亚马逊开发）。如今，Theano和CNTK都已不再开发，MXNet在亚马逊以外没有广泛使用。Keras又回到了单后端API——在TensorFlow之上。

Keras和TensorFlow多年来一直保持共生关系。在整个2016年和2017年，Keras以开发TensorFlow应用程序的用户友好方式而闻名，将新用户引导到TensorFlow生态系统中。到2017年底，大多数TensorFlow用户通过Keras或与Keras结合使用它。2018年，TensorFlow领导层选择Keras作为TensorFlow的官方高级API。因此，Keras API是2019年9月发布的TensorFlow 2.0的前沿和中心——这是TensorFlow和Keras的广泛重新设计，考虑到了四年多的用户反馈和技术进步。

````python
import tensorflow as tf
x = tf.ones(shape=(2, 1))
print(x)
````

### Tensor operations: doing math in TensorFlow

就像NumPy一样，TensorFlow提供了大量张量运算来表达数学公式。以下是一些例子：

````python
a = tf.ones((2, 2))
b = tf.square(a)
c = tf.sqrt(a)
d = b + c
e = tf.matmul(a, b)
e *= d
````

重要的是，我们刚刚编写的每个操作都是实时执行的：您可以随时打印当前结果，就像在NumPy中一样。我们称之为 `eager` 模式。

###  3.5.3 A second look at the GradientTape API

到目前为止，TensorFlow看起来很像NumPy。但NumPy做不到的事情：检索任何可微表达式相对于其输入的梯度。只需打开 `GradientTape` scope ，对一个或多个输入张量应用一些计算，并检索结果相对于输入的梯度。

````python
input_var = tf.Variable(initial_value=3.)
with tf.GradientTape() as tape:
    result = tf.square(input_var)
gradient = tape.gradient(result, input_var)
````

这最常用于检索模型权重对于损失的梯度：`gradient = tape.gradient(loss, weights)`。

在第2章中，您已经看到了 `GradientTape` 如何在单个输入或输入列表中工作，以及输入如何可以是标量或高维张量。

到目前为止，您只看到 `tape.gradient()` 中的输入张量是 TensorFlow 变量的情况。这些输入实际上可能是任何任意张量。然而，默认情况下只跟踪可训练变量。当使用常量张量时，您必须在张量上调用 `tape.watch()`，手动将其标记为被跟踪：

````python
input_const = tf.constant(3.)
with tf.GradientTape() as tape:
    tape.watch(input_const)
    result = tf.square(input_const)
gradient = tape.gradient(result, input_const)
````

为什么？因为提前存储计算任何梯度所需的信息太昂贵了。为了避免浪费资源，tape 需要知道该看什么。默认情况下会观察可训练变量，因为计算可训练变量列表的损失梯度是 gradient tape 最常见的用例。

gradient tape 是一个强大的实用工具，甚至可以计算二阶梯度，也就是说， 梯度的梯度。 例如，物体相时间位置的梯度是该物体的速度，二阶梯度是其加速度。如果您随着时间的推移测量掉落苹果沿着垂直轴的位置，并发现它验证了 `position(time) = 4.9 * time ** 2`，它的加速度是多少？让我们用双重嵌套 gradient tape 来找出答案。

````python
time = tf.Variable(0.)
with tf.GradientTape() as outer_tape:
    with tf.GradientTape() as inner_tape:
        position = 4.9 * time ** 2
    speed = inner_tape.gradient(position, time)
acceleration = outer_tape.gradient(speed, time)
````

### 3.5.4 An end-to-end example: a linear classifier in pure TensorFlow

你知道张量、变量、张量运算，也知道如何计算梯度。这足以构建任何基于梯度下降的机器学习模型。而你当前仅在第3章！

在机器学习工作面试中，您可能会被要求在TensorFlow中从头开始实现线性分类器：一项非常简单的任务，可以作为机器学习背景最小的候选人和没有机器学习背景的候选人之间的过滤器。让我们通过该过滤器，并使用您新获得的TensorFlow知识来实现这样的线性分类器。

首先，让我们想出一些线性分离良好的合成数据来处理：二维平面中的两类点。

````python
num_samples_per_class = 1000
negative_samples = np.random.multivariate_normal(
    mean=[0, 3], cov=[[1, 0.5], [0.5, 1]], size=num_samples_per_class)
positive_samples = np.random.multivariate_normal(
    mean=[3, 0], cov=[[1, 0.5], [0.5, 1]], size=num_samples_per_class)
````

`negative_samples` 和 `positive_samples` 都是形状为（1000，2）的数组。让我们把它们堆叠成一个形状（2000，2）的数组：

````python
inputs = np.vstack((negative_samples, positive_samples)).astype(np.float32)
````

让我们生成相应的目标标签，一个形状为 （2000，1）的全零数组和全一数组，其中 `targets[i, 0] is 0 if inputs[i] belongs to class 0`， 反之亦然：

````python
targets = np.vstack((np.zeros((num_samples_per_class, 1), dtype='float32'),
                     np.ones((num_samples_per_class, 1), dtype='float32')))
````

让我们使用著名的Python数据可视化库Matplotlib绘制我们的数据（它预装在Colab中，因此无需您自己安装）：

````python
import matplotlib.pyplot as plt
plt.scatter(inputs[:, 0], inputs[:, 1], c=targets[:, 0])
plt.show()
````

现在，让我们创建一个线性分类器，可以学习分离这两种点。线性分类器是一种仿射变换（`prediction = W · input + b`），训练以最小化预测和目标之间的平方差异。

如您所见，这实际上是一个比第2章末尾玩具双层神经网络的端到端示例简单得多的例子。然而， 这次， 你应该能够逐行理解代码中的一切。

让我们创建我们的变量 W 和 b，分别用随机值和零初始化：

````python
input_dim = 2
output_dim = 1
W = tf.Variable(initial_value=tf.random.uniform(shape=(input_dim, output_dim)))
b = tf.Variable(initial_value=tf.zeros(shape=(output_dim)))
````

以下是我们的前向传播函数：

````python
def model(inputs):
    return tf.matmul(inputs, W) + b
````

由于我们的线性分类器在二维输入上运行，W实际上只是两个标量系数，`w1` 和 `w2：W = [[w1]，[w2]]` 。同时，b是一个单标量系数。因此，对于给定的输入点 `[x，y]`，其预测值是：`prediction=[[w1]，[w2]] ·[x，y] + b = w1 * x + w2 * y + b`。

这里是我们的损失函数：

````python
def square_loss(targets, predictions):
    per_sample_losses = tf.square(targets - predictions)
    return tf.reduce_mean(per_sample_losses)
````

现在，训练步骤接收一些训练数据，并更新权重W和b，以尽量减少数据损失：

````python
learning_rate = 0.1

def training_step(inputs, targets):
    with tf.GradientTape() as tape:
        predictions = model(inputs)
        loss = square_loss(predictions, targets)
    grad_loss_wrt_W, grad_loss_wrt_b = tape.gradient(loss, [W, b])
    W.assign_sub(grad_loss_wrt_W * learning_rate)
    b.assign_sub(grad_loss_wrt_b * learning_rate)
    return loss
````

为了简单起见，我们将进行批处理训练，而不是小批量处理培训：我们将对整个数据运行每个训练步骤（梯度计算和权重更新），而不是以小批量迭代数据。一方面，这意味着每个训练步骤都需要更长的时间才能运行，因为我们同时计算了2000个样本的前向传播和梯度。另一方面，每次梯度更新都会更有效地减少训练数据的损失，因为它将包含来自所有训练样本的信息，而不仅仅是128个随机样本。因此，我们需要的训练步骤要少得多，我们应该使用比我们通常用于小批量处理训练的学习率更高的学习率（我们将使用上面定义的learning_rate = 0.1）。

````python
for step in range(20):
    loss = training_step(inputs, targets)
    print('Loss at step: %d: %.4f' % (step, loss))
````

30步后，训练损失似乎稳定在0.025左右。让我们绘制线性模型如何对训练数据点进行分类。由于我们的目标是零和一，如果给定的预测值低于0.5，则将其归类为“0”，如果预测值高于0.5，则归类为“1”：

````python
predictions = model(inputs)
plt.scatter(inputs[:, 0], inputs[:, 1], c=predictions[:, 0] > 0.5)
plt.show()
````

回想一下，给定点`[x, y]` 的预测值只是 `prediction == [[w1], [w2]] • [x, y] + b == w1 * x + w2 * y + b`。因此，“0”类被定义为：`w1 * x + w2 * y + b < 0.5`，类“1”被定义为：`w1 * x + w2 * y + b > 0.5`。您会注意到，您正在查看的实际上是2D平面上一条线的方程：`w1 * x + w2 * y + b = 0.5`。行上方，1类，行下方，0类。您可以习惯于查看 `y = a * x + b` 格式的线性方程；在同一格式下，我们的线变成了：`y = - w1 / w2 * x + (0.5 - b) / w2`。

让我们画出这条线：

````python
x = np.linspace(-1, 4, 100)
y = - W[0] / W[1] * x + (0.5 - b) / W[1]
plt.plot(x, y, '-r')
plt.scatter(inputs[:, 0], inputs[:, 1], c=predictions[:, 0] > 0.5)
````

这就是线性分类器的真正意义所在：找到一条线的参数（或在高维空间中，找到一个超平面），将两类数据整齐地分开。

### 3.6.1 Layers: the building blocks of deep learning

神经网络的基本数据结构是第2章中向您介绍的层。层是一个数据处理模块，以一个或多个张量作为输入，并输出一个或多个张量。一些层是无状态的，但更常见的是层具有状态：层的权重，通过随机梯度下降学习的一个或多个张量，它们一起包含网络的知识。

不同类型的层适用于不同的张量格式和不同类型的数据处理。例如，存储在形状的二维张量（样本、特征）中的简单矢量数据通常由 densely connected 的层处理，也称为全连接或 dense 的层（Keras中的 `Dense` 类）。序列数据存储在形状的3D张量（样本、时间步、特征）中，通常由循环层（如LSTM层）或1D卷积层（Conv1D）处理。存储在4D张量中的图像数据通常由2D卷积层（Conv2D）处理。

你可以把层想象成深度学习的乐高积木，Keras明确了这个比喻。Keras中建立深度学习模型通过将层剪在一起形成有用的数据变换流程。

**THE BASE LAYER CLASS IN KERAS**

一个简单的API应该有一个以所有东西为中心的抽象。在Keras中，这是 `Layer` 类。Keras中的所有内容要么是 `Layer`，要么是与 `Layer` 密切交互的东西。

`Layer` 是一个封装某些状态（权重）和一些计算（前向传播）的对象。权重通常在 `build()` 中定义（尽管它们也可以在构造函数 `init()` 中创建），计算在`call()` 方法中定义。

在上一章中，我们实现了包含两个权重 W 和 b 的 `NaiveDense` 类，并应用了计算 `output = activation(dot(W，input）+b)`。这就是Keras中同一层的样子：

````python
from tensorflow import keras

class SimpleDense(keras.layers.Layer):
    
    def __init__(self, units, activation=None):
        super(SimpleDense, self).__init__()
        self.units = units
        self.activation = activation
    
    def build(self, input_shape):
        input_dim = input_shape[-1]
        self.W = self.add_weight(shape=(input_dim, self.units),
                                 initializer='random_normal')
        self.b = self.add_weight(shape=(self.units,),
                                 initializer='zeros')
    def call(self, inputs):
        y = tf.matmul(inputs, self.W) + self.b
        if self.activation is not None:
            y = self.activation(y)
        return y
````

- 所有Keras layers 都从基类 `Layer` 继承。
- 权重创建在build()方法中进行。
- Add_weight是创建权重的快捷方式。也可以创建独立的变量并将其分配为层属性，例如：self.W = tf.Variable(tf.random.uniform(w_shape))。
- 我们在call()方法中定义了正向传播计算。

在下一节中，我们将详细介绍这些build()和call()方法的目的。

如果你现在还不明白一切，别担心！

一旦实例化，像这样的层可以像函数一样使用，以 TensorFlow 张量作为输入：

````python
my_dense = SimpleDense(units=32, activation=tf.nn.relu)
input_tensor = tf.ones(shape=(2, 784))
output_tensor = my_dense(input_tensor)
print(output_tensor.shape)
````

现在，您可能想知道，为什么我们必须实现call()和build()，因为我们最终通过简单地调用它来使用我们的层，也就是说，通过使用它的 `call` 方法？这是因为我们希望能够及时创建状态。让我们看看这是怎么回事。

**AUTOMATIC SHAPE INFERENCE: BUILDING LAYERS ON THE FLY**

就像乐高积木一样，您只能将兼容的层“夹在一起”。这里的层兼容性概念具体是指每个图层只接受特定形状的输入张量，并将返回特定形状的输出张量。考虑以下示例：

````python
from tensorflow.keras import layers
layer = layers.Dense(32, activation='relu')
````

该层将返回一个张量，其中第一个维度已转换为32。它只能连接到期望32维矢量作为输入的下游层。

使用Keras时，您大多数时候不必担心大小兼容性，因为您添加到模型中的图层是动态构建的，以匹配传入层的形状。例如，假设您写了以下内容：

````python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential([
    layers.Dense(32, activation='relu'),
    layers.Dense(32)
])
````

这些层没有收到任何关于其输入形状的信息——相反，它们会自动推断其输入形状是他们看到的第一批输入的形状。

在我们在第2章中实现的 `Dense` 层的玩具版本（我们将其命名为 `NaiveDense` ）中，我们必须将该层的输入大小显式传递给构造函数，以便能够创建其权重。这并不理想，因为它会导致看起来像这样的模型，每个新层都需要在它之前了解层的形状：

```python
model = NaiveSequential([
  NaiveDense(input_size=784, output_size=32, activation='relu'),
  NaiveDense(input_size=32, output_size=64, activation='relu'),
  NaiveDense(input_size=64, output_size=32, activation='relu'),
  NaiveDense(input_size=32, output_size=10, activation='softmax')
])
```

当层用于产生输出形状的规则复杂时，情况会更糟。例如，如果我们的层返回形状的输出`(batch，input_size * 2 if input_size % 2 == 0 else input_size * 3)`，该怎么办？

如果我们将我们的 `NaiveDense` 层重新实现为能够自动形状推断的Keras层，它将更简单地看起来像上面的 `SimpleDense` 层 ，及其 `build()` 和 `call()` 方法。

在 `SimpleDense` 中，我们不再像 `NaiveDense` 示例那样在构造函数中创建权重，而是在专用状态创建方法 `build()` 中创建权重，该方法将层看到的第一个输入形状作为参数接收。`build()` 方法在首次调用该层时自动调用（通过其 `call` 方法）。事实上，这就是为什么我们在 `call` 方法中定义计算！基类 layer 的 `call` 方法原理如下：

```python
def __call__(self, inputs):
    if not self.built:
        self.build(inputs.shape)
        self.built = True
    return self.call(inputs)
```

通过自动形状推断，我们之前的例子变得简单整洁：

````python
model = keras.Sequential([
    SimpleDense(32, activation='relu'),
    SimpleDense(64, activation='relu'),
    SimpleDense(32, activation='relu'),
    SimpleDense(10, activation='softmax'),
])
````

请注意，自动形状推断不是 `Layer` 类 `call` 方法处理的唯一问题。它处理了许多其他事情，特别是在 `eager` 执行和 `graph` 执行（您将在第6章中了解的概念）和输入 masking 。现在，请记住：在实现自己的层时，请在 `call` 方法中提供正向传播。

### 3.6.4 Picking a loss function

为正确的问题选择正确的损失函数极其重要：您的网络将采取任何捷径来尽量减少损失；因此，如果目标与手头任务的成功不完全相关，您的网络最终将做您可能不想要的事情。想象一下，一个通过SGD训练的愚蠢、无所不能的人工智能，其目标功能选择不当：“最大限度地提高所有活着的人的平均福祉。”为了使其工作更容易，除了少数人外，这种人工智能可能会选择杀死所有人类，并专注于其余人的福祉——因为平均福祉不受剩余人数的影响。那可能不是你的意图！请记住，您构建的所有神经网络在降低其损失函数时都会同样无情——因此请明智地选择目标，否则您将不得不面对意外的副作用。

幸运的是，当涉及到分类、回归和序列预测等常见问题时，您可以遵循简单的准则来选择正确的损失。例如，您将对二分类问题使用 binary crossentropy，对多类分类问题使用 categorical crossentropy，等等。只有当你处理真正新的研究问题时，你才必须开发自己的目标功能。在接下来的几章中，我们将明确详细说明为广泛的常见任务选择哪些损失函数。

### 3.6.6 Monitoring loss & metrics on validation data

机器学习的目标不是获得在训练数据上表现良好的模型——这很简单，你所要做的就是遵循梯度。目标是获得总体性能良好的模型，特别是在模型以前从未遇到过的数据点上。仅仅因为模型在训练数据上表现良好并不意味着它将在从未见过的数据上表现良好！例如，您的模型最终可能只是记住训练样本及其目标之间的映射，这对预测模型以前从未见过的数据的目标任务毫无用处。我们将在第5章中更详细地讨论这一点。

为了关注模型在新数据上的表现，标准做法是将训练数据的子集保留为“验证数据”：您不会根据这些数据训练模型，但您将使用它来计算损失值和指标值。您可以使用fit()中的 `validation_data` 参数来完成此操作。与训练数据一样，验证数据可以作为 NumPy 数组或 TensorFlow 数据集对象传递。

````python
model = keras.Sequential([keras.layers.Dense(1)])
model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=0.1),
              loss=keras.losses.MeanSquaredError(),
              metrics=[keras.metrics.BinaryAccuracy()])

indices_permutation = np.random.permutation(len(inputs))
shuffled_inputs = inputs[indices_permutation]
shuffled_targets = targets[indices_permutation]

num_validation_samples = int(0.3 * len(inputs))
val_inputs = shuffled_inputs[-num_validation_samples:]
val_targets = shuffled_targets[-num_validation_samples:]
training_inputs = shuffled_inputs[:-num_validation_samples]
training_targets = shuffled_targets[:-num_validation_samples]

model.fit(
    training_inputs,
    training_targets,
    epochs=5,
    batch_size=16,
    validation_data=(val_inputs, val_targets)
)
````

- 为了避免验证数据中只有一个类的样本，请使用随机索引排列来洗牌输入和目标
- 保留20%的训练样本和目标用于“验证”（我们将把这些样本排除在训练之外，并保留它们来计算“验证损失”和指标）
- 训练数据，用于更新模型的权重
- 验证数据，仅用于监控“验证损失”和指标

验证数据上的损失值称为“验证损失”，以将其与“训练损失”区分开来。请注意，必须严格将训练数据和验证数据分开：验证的目的是监控模型正在学习的内容是否真的对新数据有用。如果模型在训练期间看到任何验证数据，您的验证损失和指标将有缺陷。

请注意，如果您想在训练完成后计算验证损失和指标，您可以调用 `evaluate` 方法：

```python
loss_and_metrics = model.evaluate(val_inputs, val_targets, batch_size=128)
```

`evaluation()` 将以批处理（batch_size）迭代传递的数据，并返回标量列表，其中第一个条目是验证损失，以下条目是验证指标。如果模型没有指标，则只返回验证损失而不是列表）。

### 3.6.7 Inference: using a model after training

一旦您训练了模型，您将希望使用它来预测新数据。这被称为“推理”。要做到这一点，一个简单的方法就是调用模型：

```python
predictions = model(new_inputs)
```

然而，这将同时处理 `new_inputs` 中的所有输入，如果您查看大量数据（特别是，它可能需要比GPU更多的内存），这可能不可行。

更好的推理方法是使用 `predict()`方法。它将以小批量迭代数据，并返回NumPy预测数组。与 `call` 不同，它还可以处理TensorFlow数据集对象。

```python
predictions = model.predict(new_inputs, batch_size=128)
```

例如，如果我们对我们之前训练的线性模型的一些验证数据使用 `predict()`，我们得到的标量分数在0到1之间——低于0.5表示模型认为对应点属于0类，高于0.5表示模型认为对应点属于第1类。


````python
predictions = model.predict(val_inputs, batch_size=128)
print(predictions[:10])
````
